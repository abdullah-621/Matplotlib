```python
df = pd.read_csv("/content/drive/MyDrive/Matplotlib_Dataset/batter.csv")
df = df.head(25)
df
```

```python
plt.figure(figsize=(20,10))
plt.scatter(df['avg'],df['strike_rate'], s = df['runs'])
plt.grid()

for i in range(df.shape[0]):
  plt.text(df['avg'].values[i], df['strike_rate'].values[i],df['batter'].values[i])
```

# Vertical and Horizental lines

```python
plt.figure(figsize=(20,10))
plt.scatter(df['avg'],df['strike_rate'], s = df['runs'])
plt.grid()
plt.axhline(130, color = "red")
plt.axhline(140, color = "yellow")
plt.axvline(30, color = "blue")

for i in range(df.shape[0]):
  plt.text(df['avg'].values[i], df['strike_rate'].values[i],df['batter'].values[i])
```

# Subplots

```python
fig, ax = plt.subplots(figsize = (18,9))

ax.scatter(df['avg'], df['runs'])
ax.set_title("Runs and Avg")
ax.set_xlabel("AVG")
ax.set_ylabel("RUNS")

for i in range(df.shape[0]):
  ax.text(df['avg'].values[i], df['runs'].values[i], df['batter'].values[i])
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, figsize = (10,8))
ax[0].scatter(df['runs'], df['avg'], color = "red")
ax[1].scatter(df['runs'], df['strike_rate'], color = "black")

ax[0].set_title("Runs and Avg")
ax[0].set_ylabel("Avg")


ax[1].set_title("Runs and strike_rate")
ax[1].set_ylabel("Strike Rate")
ax[1].set_xlabel("Runs")
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (10,8))
ax[0,0].scatter(df['runs'], df['avg'], color = "red")
ax[0,1].scatter(df['runs'], df['strike_rate'], color = "black")
ax[1,0].hist(df['avg'], bins = 10)
ax[1,1].hist(df['strike_rate'], bins = 10)
fig.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(1,1,1)
ax1.scatter(df['runs'], df['avg'], color = "red")

ax1 = fig.add_subplot(1,2,1)
ax1.scatter(df['runs'], df['avg'], color = "red")
```

# 3D Scatter plot

```python
fig = plt.figure()

ax = plt.subplot(projection = "3d")
ax.scatter3D(df['avg'], df['runs'], df['strike_rate'])

ax.set_title("IPL Batsman Analysis")

ax.set_xlabel("AVG")
ax.set_ylabel("Runs")
ax.set_zlabel("Strike Rate")
```

# **3d Surface Plot**

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx , yy = np.meshgrid(x , y)

z = xx ** 2 + yy ** 2
```

```python
fig = plt.figure(figsize = (16, 10))

ax = plt.subplot(projection = "3d")
p = ax.plot_surface(xx, yy, z, cmap = "viridis")
fig.colorbar(p)

ax.set_title("3D Surface: z = x + y")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
```

```python
fig = plt.figure(figsize = (16, 10))

ax = plt.subplot()
p = ax.contourf(xx, yy, z, cmap = "viridis")
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize = (10,6))

ax = plt.subplot(projection = "3d")
p = ax.plot_surface(xx, yy, z, cmap = "viridis")
fig.colorbar(p)
```

# Countour Plot

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize = (10,6))

ax = plt.subplot()
p = ax.contour(xx, yy, z, cmap = "viridis")
fig.colorbar(p)
```
