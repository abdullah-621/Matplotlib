```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv("/content/drive/MyDrive/Matplotlib_Dataset/iris.csv")
iris.sample(5)
```

```python
pd.set_option('future.no_silent_downcasting', True)
iris['Species'] = iris['Species'].replace({"Iris-setosa" : 0, "Iris-versicolor" : 1, "Iris-virginica" : 2})
iris.sample(5)
```

```python
# cmap, alpha , plot size

plt.figure(figsize=(16,8))
plt.scatter(iris['PetalLengthCm'],iris['SepalWidthCm'], c = iris['Species'], cmap = "jet", alpha = 0.9)
plt.xlabel("Petal Length")
plt.ylabel("Sepal Width")
plt.colorbar()
```
