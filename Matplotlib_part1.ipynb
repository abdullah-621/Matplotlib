```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# ***2D Ploting***

```python
# 2d ploting

price = [300000, 450000, 5000000, 6000000, 7000000]
year = [2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)
```

```python
batsman = pd.read_csv("/content/drive/MyDrive/Matplotlib_Dataset/Matplotlib/datasets-session-23/sharma-kohli.csv")
batsman
```

```python
# color , line , marker , legend

plt.plot(batsman['index'], batsman['RG Sharma'], color = "Red", label = "RG Sharma")
plt.plot(batsman['index'], batsman['V Kohli'], color = "Blue", label = "V Kohli")

plt.title('V Kohli & Rohit Sharma IPL Career Comparison')
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()  # To show labels
# plt.legend(loc = 'upper right')
# plt.show()
```

```python
# from matplotlib.lines import lineStyles

plt.plot(batsman['index'], batsman['RG Sharma'], color = "Red", label = "RG Sharma", linestyle = "solid", linewidth = 3)
plt.plot(batsman['index'], batsman['V Kohli'], color = "Blue", label = "V Kohli", linestyle = 'dotted', linewidth = 4)

plt.title('V Kohli & Rohit Sharma IPL Career Comparison')
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()  # To show labels
```

```python
plt.plot(batsman['index'], batsman['RG Sharma'], color = "Red", label = "RG Sharma", marker = 'o', markersize = 10)
plt.plot(batsman['index'], batsman['V Kohli'], color = "Blue", label = "V Kohli", marker = '+', markersize = 8)

plt.title('V Kohli & Rohit Sharma IPL Career Comparison')
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()  # To show labels
```

```python
# limiting axis
price = [300000, 450000, 5000000, 6000000, 7000000, 800000000]
year = [2016, 2017, 2018, 2019, 2020, 2021]

plt.plot(year, price)
plt.ylim(0, 6000000)
plt.xlim(2017, 2020)
```

```python
# grid

plt.plot(batsman['index'], batsman['RG Sharma'], color = "Red", label = "RG Sharma", marker = 'o', markersize = 10)
plt.plot(batsman['index'], batsman['V Kohli'], color = "Blue", label = "V Kohli", marker = '+', markersize = 8)

plt.title('V Kohli & Rohit Sharma IPL Career Comparison')
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()  # To show labels
plt.grid()
```

```python
# show

plt.plot(batsman['index'], batsman['RG Sharma'], color = "Red", label = "RG Sharma", marker = 'o', markersize = 10)
plt.plot(batsman['index'], batsman['V Kohli'], color = "Blue", label = "V Kohli", marker = '+', markersize = 8)

plt.title('V Kohli & Rohit Sharma IPL Career Comparison')
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()  # To show labels
plt.grid()
plt.show()
```
