```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# ***2D Ploting***

```python
# 2d ploting

price = [300000, 450000, 5000000, 6000000, 7000000]
year = [2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)
```

```python
batsman = pd.read_csv("/content/drive/MyDrive/Matplotlib_Dataset/Matplotlib/datasets-session-23/sharma-kohli.csv")
batsman
```

```python
# color , line , marker , legend

plt.plot(batsman['index'], batsman['RG Sharma'], color = "Red", label = "RG Sharma")
plt.plot(batsman['index'], batsman['V Kohli'], color = "Blue", label = "V Kohli")

plt.title('V Kohli & Rohit Sharma IPL Career Comparison')
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()  # To show labels
# plt.legend(loc = 'upper right')
# plt.show()
```

```python
# from matplotlib.lines import lineStyles

plt.plot(batsman['index'], batsman['RG Sharma'], color = "Red", label = "RG Sharma", linestyle = "solid", linewidth = 3)
plt.plot(batsman['index'], batsman['V Kohli'], color = "Blue", label = "V Kohli", linestyle = 'dotted', linewidth = 4)

plt.title('V Kohli & Rohit Sharma IPL Career Comparison')
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()  # To show labels
```

```python
plt.plot(batsman['index'], batsman['RG Sharma'], color = "Red", label = "RG Sharma", marker = 'o', markersize = 10)
plt.plot(batsman['index'], batsman['V Kohli'], color = "Blue", label = "V Kohli", marker = '+', markersize = 8)

plt.title('V Kohli & Rohit Sharma IPL Career Comparison')
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()  # To show labels
```

```python
# limiting axis
price = [300000, 450000, 5000000, 6000000, 7000000, 800000000]
year = [2016, 2017, 2018, 2019, 2020, 2021]

plt.plot(year, price)
plt.ylim(0, 6000000)
plt.xlim(2017, 2020)
```

```python
# grid

plt.plot(batsman['index'], batsman['RG Sharma'], color = "Red", label = "RG Sharma", marker = 'o', markersize = 10)
plt.plot(batsman['index'], batsman['V Kohli'], color = "Blue", label = "V Kohli", marker = '+', markersize = 8)

plt.title('V Kohli & Rohit Sharma IPL Career Comparison')
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()  # To show labels
plt.grid()
```

```python
# show

plt.plot(batsman['index'], batsman['RG Sharma'], color = "Red", label = "RG Sharma", marker = 'o', markersize = 10)
plt.plot(batsman['index'], batsman['V Kohli'], color = "Blue", label = "V Kohli", marker = '+', markersize = 8)

plt.title('V Kohli & Rohit Sharma IPL Career Comparison')
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()  # To show labels
plt.grid()
plt.show()
```

```python
# simple bar chart

children = [10,20,40,10,30]
colors = ['red', 'green', 'blue', 'yellow', 'pink']

plt.bar(colors, children, color = 'gray')
plt.title('Children like color analysis')
plt.xlabel("colors")
plt.ylabel("children")
```

```python
df = pd.read_csv("/content/drive/MyDrive/Matplotlib_Dataset/Matplotlib/datasets-session-23/batsman_season_record.csv")
df
```

```python
plt.bar(df['batsman'], df['2015'])
plt.title("Top 5 batsman run in 2015")
plt.xlabel("Batsman")
plt.ylabel("Runs")
```

```python

plt.bar(df['batsman'], df['2015'], width = 0.8, label = "2015")
plt.bar(df['batsman'], df['2016'], width = 0.5, label = "2016")
plt.bar(df['batsman'], df['2017'], width = 0.3, label = "2017")
plt.legend()
```

```python
# multiple bar chart
plt.bar(np.arange(df.shape[0]) + 0.2, df['2015'], width = 0.2, label = '2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width = 0.2, label = '2016')
plt.bar(np.arange(df.shape[0]) - 0.2, df['2017'], width = 0.2, label = '2017')
plt.legend()
plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

```python
# problem (big name of catagorical data)
children = [10,20,40,10,30]
colors = ['red red red red', 'green green green green green', ' blue  blue blue blue blue', ' yellow yellow yellow yellow', ' pink pink pink']

plt.bar(colors, children, color = 'gray')
plt.xticks(rotation = "vertical")
```

```python
# stacked bar chart
plt.bar(df['batsman'], df['2015'], label = "2015")
plt.bar(df['batsman'], df['2016'], label = "2016", bottom = df['2015'])
plt.bar(df['batsman'], df['2017'], label = "2017", bottom = df['2015'] + df['2016'])
plt.legend()
```

# pie chart

```python
number = [100, 30, 40, 79, 59]
subjects = ['eng', 'bng', 'math', 'science', 'sta']

plt.pie(number, labels = subjects, autopct = "%0.1f%%")
plt.show()
```

```python
# dataset

df = pd.read_csv("/content/drive/MyDrive/Matplotlib_Dataset/Matplotlib/datasets-session-23/gayle-175.csv")
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = "%0.1f%%", colors = ['blue', 'green','red', 'black', 'purple', 'skyblue'])
plt.show()
```
